```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


df = pd.read_csv("healthcare_dataset.csv")


print("Rows, Columns:", df.shape)
print(df.dtypes)
print("\nSample rows:")
display(df.head())


drop_cols = [
    "Name",
    "Doctor",
    "Hospital",
    "Date of Admission",
    "Discharge Date",
    "Insurance Provider"
]
df_model = df.drop(columns=drop_cols)
print("\nColumns used for modeling:", df_model.columns.tolist())


target_col = "Test Results"
X = df_model.drop(columns=[target_col]).copy()
y = df_model[target_col].copy()


encoders = {}
for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        encoders[col] = le


target_le = LabelEncoder()
y_encoded = target_le.fit_transform(y.astype(str))

print("\nExample encoded feature columns (first 5 rows):")
display(X.head())


X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.20, random_state=42, stratify=y_encoded
)
print(f"\nTrain size: {X_train.shape[0]} rows, Test size: {X_test.shape[0]} rows")


rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)


y_pred = rf.predict(X_test)


acc = accuracy_score(y_test, y_pred)
print(f"\nAccuracy on test set: {acc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_le.classes_))


cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=target_le.classes_, yticklabels=target_le.classes_, cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix  Test Results")
plt.show()


comparison = X_test.copy()
comparison = comparison.reset_index(drop=True)
comparison["Actual_encoded"] = y_test
comparison["Pred_encoded"] = y_pred
comparison["Actual"] = target_le.inverse_transform(comparison["Actual_encoded"])
comparison["Predicted"] = target_le.inverse_transform(comparison["Pred_encoded"])


for col, le in encoders.items():

    try:
        comparison[col + "_label"] = le.inverse_transform(comparison[col].astype(int))
    except Exception:

        comparison[col + "_label"] = comparison[col]


pd.set_option('display.max_columns', None)
display(comparison[[c for c in comparison.columns if "_label" in c][:6]
                   ] .join(comparison[["Actual", "Predicted"]]).head(20))


display(comparison[["Actual", "Predicted"]].head(30))

fi = pd.DataFrame({
    "feature": X.columns,
    "importance": rf.feature_importances_
}).sort_values("importance", ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(data=fi.head(12), x="importance", y="feature")
plt.title("Top 12 Feature Importances (Random Forest)")
plt.show()

```

```python

!pip install xgboost > /dev/null

import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
import seaborn as sns, matplotlib.pyplot as plt


df = pd.read_csv("healthcare_dataset.csv")


df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Stay_Days'] = (df['Discharge Date'] - df['Date of Admission']).dt.days.fillna(0).astype(int)
df['Billing_bucket'] = pd.qcut(df['Billing Amount'].clip(lower=0), 5, labels=False)


cond_cancer = df['Medical Condition'].str.contains('Cancer', case=False, na=False)
high_bill = df['Billing_bucket'] == df['Billing_bucket'].max()
age_over60 = df['Age'] > 60
med_list = df['Medication'].str.lower().isin(['ibuprofen','aspirin','penicillin'])

labels = np.where(cond_cancer | high_bill, 2, np.where(age_over60 & med_list, 1, 0))
rng = np.random.default_rng(42)
flip_idx = rng.choice(len(labels), size=int(len(labels)*0.12), replace=False)
labels_noisy = labels.copy()
labels_noisy[flip_idx] = rng.integers(0, 3, size=flip_idx.shape[0])
map_lbl = {0:'Normal', 1:'Inconclusive', 2:'Abnormal'}
df['TestResults_synthetic'] = [map_lbl[x] for x in labels_noisy]

print("Synthetic target created. Class distribution:")
print(df['TestResults_synthetic'].value_counts().to_dict())


X = df[['Age','Billing Amount','Room Number','Stay_Days','Billing_bucket']].copy()
X['Gender_code'] = df['Gender'].astype('category').cat.codes
X['MedCond_code'] = df['Medical Condition'].astype('category').cat.codes
X['Medication_code'] = df['Medication'].astype('category').cat.codes


y = df['TestResults_synthetic'].astype(str)
label_enc = LabelEncoder()
y_enc = label_enc.fit_transform(y)
print("Label mapping (int -> label):", dict(enumerate(label_enc.classes_)))


X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.20, random_state=42, stratify=y_enc)


model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)


model.fit(X_train, y_train)


y_pred_enc = model.predict(X_test)


y_test_labels = label_enc.inverse_transform(y_test)
y_pred_labels = label_enc.inverse_transform(y_pred_enc)


acc = accuracy_score(y_test_labels, y_pred_labels)
print(f"\nSYNTHETIC TARGET ACCURACY: {acc:.4f}\n")
print("Classification Report:")
print(classification_report(y_test_labels, y_pred_labels))


cm = confusion_matrix(y_test_labels, y_pred_labels, labels=label_enc.classes_)
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=label_enc.classes_, yticklabels=label_enc.classes_, cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Synthetic Target)")
plt.show()
```

```python


import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("healthcare_dataset.csv")


df["Billing_ZScore"] = (df["Billing Amount"] - df["Billing Amount"].mean()) / df["Billing Amount"].std()


df["Anomaly_ZScore"] = df["Billing_ZScore"].abs() > 3


iso = IsolationForest(contamination=0.02, random_state=42)  # 2% anomalies
df["Anomaly_IF"] = iso.fit_predict(df[["Billing Amount"]])


df["Anomaly_IF"] = df["Anomaly_IF"].apply(lambda x: True if x == -1 else False)


df["Final_Anomaly"] = df["Anomaly_ZScore"] | df["Anomaly_IF"]


anomalies = df[df["Final_Anomaly"]]
print("Total anomalies detected:", len(anomalies))

display(anomalies.head(10))


plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x=df.index, y="Billing Amount", hue="Final_Anomaly", palette=["blue","red"])
plt.title("Billing Amount Anomalies (Red = Anomaly)")
plt.xlabel("Index")
plt.ylabel("Billing Amount")
plt.show()


print("\nAnomaly Billing Range:")
print("Min:", anomalies["Billing Amount"].min())
print("Max:", anomalies["Billing Amount"].max())

print("\nTypical Billing Range (Non-Anomalies):")
normal = df[~df["Final_Anomaly"]]
print("Min:", normal["Billing Amount"].min())
print("Max:", normal["Billing Amount"].max())
```
